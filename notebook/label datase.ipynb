```python
import pandas as pd
import re

# Load your DataFrame 
df = pd.read_csv("data/clean/messages_preprocessed.csv")

def simple_tokenizer(text):
    # Split based on space and preserve punctuation
    tokens = re.findall(r'\w+|[፣፡።,.:!?]', text)
    return tokens

def label_token(token):
    # Rules for basic labeling
    if re.fullmatch(r'\d+', token):
        return 'I-PRICE'
    elif 'ብር' in token:
        return 'I-PRICE'
    elif token in ['ዋጋ', 'በ', 'ከ']:
        return 'B-PRICE'
    elif token in ['አዲስ', 'ቦሌ', 'አካባቢ', 'ሰሚት', 'መከኒሳ', 'ጉልላሌ', 'አያት', 'ሽሮ ምዳ', 'አራት ኪሎ', 'አምስት ኪሎ', 'ስድስት ኪሎ', 'መርካቶ', 'አዲስ ከተማ']:  # Add known locations here
        return 'B-LOC'
    elif token in ["መኪና", "መሳሪያ", "ሶፋ", "ታቦት", "እቃ", "የህጻናት",
    "መክብብ", "አልባሳት", "ቀሚስ", "ብርቱካና", "አልጋ", "መድረክ",
    "ጫማ", "ቦት", "ሳል", "መጠጥ", "እንቁላል", "አንጀት", "ሸማኔ", 
    "ካሜራ", "ቤት", "ማንኛውም", "የቤት", "የመኪና", "በይነመረብ", 
    "መጻሕፍት", "መቀመጫ", "ኮምፒውተር", "ሎሌቶ", "ኢንተርኔት", "ተሽከርካሪ",
    "የህንፃ", "ችላ", "ሻምፓን", "ሽቶ", "እርስተላላኪ", "ፋልኮን", "ከባድ ምርቶች"
    ]:  # Add known product starts here
        return 'B-Product'
    elif token in ['ጠባቂ', 'ቤት']:  # Continuation of product name
        return 'I-Product'
    else:
        return 'O'

def label_message(message):
    tokens = simple_tokenizer(message)
    labeled = [(tok, label_token(tok)) for tok in tokens]
    return labeled

def write_conll_format(labeled_messages, output_file='labeled_data.conll'):
    with open(output_file, 'w', encoding='utf-8') as f:
        for message in labeled_messages:
            for token, label in message:
                f.write(f"{token} {label}\n")
            f.write("\n")  # Separate messages

# Example usage
def process_dataframe(df):
    labeled_data = []
    for message in df["tokens"]:
        labeled = label_message(message)
        labeled_data.append(labeled)
    write_conll_format(labeled_data)
    print("Saved labeled data to labeled_data.conll")

# Replace df below with your actual DataFrame

# process_dataframe(df)
```
